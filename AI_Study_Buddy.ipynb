```python
import google.generativeai as genai
import gradio as gr

# Paste your API key here
GEMINI_API_KEY = "Your api "

genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel("gemini-2.5-flash")

print("‚úÖ Gemini API connected successfully!")
```

```python
def explain_topic(topic, level):
    prompt = f"""
    Explain the topic: "{topic}"
    Target audience: {level} student

    Format your response as:
    üìñ SIMPLE EXPLANATION:
    (2-3 sentences, very easy to understand)

    üîë KEY POINTS:
    (3-5 bullet points)

    üí° REAL LIFE EXAMPLE:
    (one relatable example)
    """
    response = model.generate_content(prompt)
    return response.text


def summarize_notes(notes):
    prompt = f"""
    Summarize these study notes in a clear, concise way:

    {notes}

    Format your response as:
    üìù SUMMARY:
    (short paragraph)

    üîë MUST REMEMBER:
    (top 5 key points as bullet points)
    """
    response = model.generate_content(prompt)
    return response.text


def generate_quiz(topic, num_questions):
    prompt = f"""
    Create {num_questions} multiple choice questions about: "{topic}"

    For each question use this exact format:

    Q1. (question here)
    A) option
    B) option
    C) option
    D) option
    ‚úÖ Answer: (correct option)

    Make questions progressively harder.
    """
    response = model.generate_content(prompt)
    return response.text


def generate_flashcards(topic, num_cards):
    prompt = f"""
    Create {num_cards} flashcards for the topic: "{topic}"

    Use this exact format for each:

    üÉè CARD (number):
    FRONT: (question or term)
    BACK: (answer or definition)
    ---
    """
    response = model.generate_content(prompt)
    return response.text
```

```python
def explain_wrapper(topic, level):
    try:
        return explain_topic(topic, level)
    except Exception as e:
        return f"Error: {str(e)}"

def summarize_wrapper(notes):
    try:
        return summarize_notes(notes)
    except Exception as e:
        return f"Error: {str(e)}"

def quiz_wrapper(topic, num):
    try:
        return generate_quiz(topic, int(num))
    except Exception as e:
        return f"Error: {str(e)}"

def flash_wrapper(topic, num):
    try:
        return generate_flashcards(topic, int(num))
    except Exception as e:
        return f"Error: {str(e)}"

with gr.Blocks(title="AI Study Buddy") as app:

    gr.Markdown("""
    # üéì AI Study Buddy
    ### Your personal AI-powered learning assistant
    """)

    with gr.Tab("üìñ Explain Topic"):
        topic_input = gr.Textbox(label="Topic", placeholder="e.g. Photosynthesis, Recursion...")
        level_input = gr.Dropdown(["School", "College", "Beginner", "Advanced"], label="Your Level", value="College")
        explain_btn = gr.Button("‚ú® Explain It!", variant="primary")
        explain_output = gr.Textbox(label="Explanation", lines=12)
        explain_btn.click(fn=explain_wrapper, inputs=[topic_input, level_input], outputs=explain_output)

    with gr.Tab("üìù Summarize Notes"):
        notes_input = gr.Textbox(label="Paste Your Notes Here", lines=10)
        summarize_btn = gr.Button("üìã Summarize!", variant="primary")
        summary_output = gr.Textbox(label="Summary", lines=12)
        summarize_btn.click(fn=summarize_wrapper, inputs=notes_input, outputs=summary_output)

    with gr.Tab("‚ùì Quiz Generator"):
        quiz_topic = gr.Textbox(label="Topic", placeholder="e.g. Python Functions...")
        num_q = gr.Slider(minimum=3, maximum=10, value=5, step=1, label="Number of Questions")
        quiz_btn = gr.Button("üß† Generate Quiz!", variant="primary")
        quiz_output = gr.Textbox(label="Your Quiz", lines=20)
        quiz_btn.click(fn=quiz_wrapper, inputs=[quiz_topic, num_q], outputs=quiz_output)

    with gr.Tab("üÉè Flashcard Generator"):
        flash_topic = gr.Textbox(label="Topic", placeholder="e.g. Data Structures...")
        num_cards = gr.Slider(minimum=3, maximum=10, value=5, step=1, label="Number of Flashcards")
        flash_btn = gr.Button("üÉè Generate Flashcards!", variant="primary")
        flash_output = gr.Textbox(label="Your Flashcards", lines=20)
        flash_btn.click(fn=flash_wrapper, inputs=[flash_topic, num_cards], outputs=flash_output)

app.launch(share=True, debug=True)
```
